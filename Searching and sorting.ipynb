```python
#Ascending selection sort
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for i in range(len(my_list)):
  minIndex = i
  for j in range(i+1,len(my_list)):
    if my_list[j] < my_list[minIndex]:
      minIndex = j
  temp = my_list[i]
  my_list[i] = my_list[minIndex]
  my_list[minIndex] = temp
print(my_list)
```

```python
#ascending bubble sort
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for i in range(len(my_list)-1):
  for j in range(len(my_list)-i-1):
    if my_list[j+1] < my_list[j]:
      temp = my_list[j]
      my_list[j] = my_list[j+1]
      my_list[j+1] = temp
print(my_list)
```

```python
#Descending selection sort
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for i in range(len(my_list)):
  maxIndex = i
  for j in range(i+1,len(my_list)):
    if my_list[j] > my_list[maxIndex]:
      maxIndex = j
  temp = my_list[i]
  my_list[i] = my_list[maxIndex]
  my_list[maxIndex] = temp
print(my_list)
```

```python
#Descending bubble sort
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for i in range(len(my_list)-1):
  for j in range(len(my_list)-i-1):
    if my_list[j+1] > my_list[j]:
      temp = my_list[j]
      my_list[j] = my_list[j+1]
      my_list[j+1] = temp
print(my_list)
```

```python
#Linear searching
def sequential_search(data, key):
    for index in range(len(data)):
         if key == data[index]:
             return index #index if found
    return -1
index = sequential_search([1,2,5,2,3],5)
if index != -1:
  print("Element Found at index",index)
else:
  print("Element not found")
```

```python
def binary_search(data, size, key):
    left = 0
    right = size - 1
    while left <= right:
        mid_index = (left+ right)//2
        mid_val = data[mid_index]
        if key == mid_val:
            return mid_index
        elif key > mid_val:
            left = mid_index + 1
        else:
            right = mid_index - 1
    return -1
data =[2,3,5,6,10,15]
index = binary_search(data,len(data),2)
if index == -1:
  print("Not found")
else:
  print("Found at location:",index)
```
