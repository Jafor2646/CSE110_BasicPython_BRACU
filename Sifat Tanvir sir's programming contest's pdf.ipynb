```python
#Super Calculator
import math
def add(n1, n2):
  return "Addition: "+ (str(n1 + n2)[:-2] if str(n1+n2)[len(str(n1+n2))-2:] == ".0" else str(n1+n2))
def sub(n1, n2):
  return "Substraction: "+ (str(n1 - n2)[:-2] if str(n1-n2)[len(str(n1-n2))-2:] == ".0" else str(n1-n2))
def mul(n1, n2):
  return "Multiplication: "+ str(n1*n2)
def divide(n1, n2):
  if n2 == 0:
    return "Invalid Input: Division by zero"
  else:
    return "Division: "+"{:.2f}".format(n1/n2)
def floor(n1, n2):
  if n2 == 0:
    return "Invalid Input: Division by zero"
  else:
    return "Floor division: "+(str(n1 // n2)[:-2] if str(n1//n2)[len(str(n1//n2))-2:] == ".0" else str(n1//n2))
def root(n1, n2):
  if n1 < 0:
    return "Invalid Input: Root of negative number"
  else:
    return "Exponentiation/Root: " + str(math.pow(n1, n2))
def mod(n1, n2):
  if n2 == 0:
    return "Invalid Input: Division by zero"
  else:
    return "Modulus: "+(str(n1 % n2)[:-2] if str(n1%n2)[len(str(n1%n2))-2:] == ".0" else str(n1%n2))
def neg(n1):
  return "Negation: "+(str(-1*n1)[:-2] if str(-1*n1)[len(str(-1*n1))-2:] == ".0" else str(-1*n1))
def compare(n1, n2):
  if n1 > n2:
    return "First Number is greater than second number"
  elif n1 < n2:
    return "First Number is less than second number"
  else:
    return "Both numbers are equal"
def state(n1):
  if n1 < 0:
    return "The number is negative"
  else:
    return "The number is positive"
op = int(input('''Which Operation you want to perform? (0=Addition,
1=Subtraction, 2=Multiplication, 3=Division, 4= Floor division,
5= Exponentiation/ Root Operation, 6= Modulus, 7= Negation,
8= Compare, 9= State of the number): '''))
if op < 0 or op > 9:
  print("Invalid operation choice")
else:
  if op == 7 or op == 9:
    num = int(input("Number: "))
    firstOperation = {7: neg(num),9: state(num)}
    print(firstOperation[op])
  else:
    fNum = float(input("First Number: "))
    sNum = float(input("Second Number: "))
    secondOperation = {0: add(fNum, sNum),1: sub(fNum, sNum),2: mul(fNum, sNum),3: divide(fNum, sNum),4: floor(fNum, sNum),5: root(fNum, sNum),6: mod(fNum, sNum),8: compare(fNum, sNum)}
    print(secondOperation[op])
```

```python
#Set Problem
def union(a,b):
    myList = []
    for i in a:
        myList.append(i)
    for i in b:
        if i not in myList:
            myList.append(i)
    return "A U B:" + str(myList)
def intersection(a,b):
    myList = []
    for i in a:
        for j in b:
            if i == j:
                if  i not in myList:
                    myList.append(i)
                    break
    return "A ∩ B:" + str(myList)
def complement(u,a):
    temp = False
    for i in a:
        if i not in u:
            temp = True
            break
    if temp:
        return "Error: First Set Containing Elements that are not a Part of the Universal Set."
    else:
        for i in a:
            u.remove(i)
        return "A’:" + str(u)
def difference(a,b):
    for i in b:
        if i not in a:
            continue
        else:
            a.remove(i)
    return "A – B:" + str(a)
def crossProduct(a,b):
    myList = []
    for i in a:
        if len(b) == 0:
            myList.append([i, None])
        else:
            for j in b:
                myList.append([i, j])
    return "A X B:" + str(myList)
setOperation = int(input("Set Operation (1=Union, 2=Intersection,3=Complement, 4=Difference, 5=Cross Product):"))
if setOperation == 3:
    li1 = input("A:")[1:-1]
    if len(li1) == 0:
        li1 = []
    else:
        li1 = li1.split(", ")
    for i in range(len(li1)):
        li1[i] = int(li1[i])
    li2 = input("Universal set, U:")[1:-1]
    if len(li2) == 0:
        li2 = []
    else:
        li2 = li2.split(", ")
    for i in range(len(li2)):
        li2[i] = int(li2[i])
else:
    li1 = input("A:")[1:-1]
    if len(li1) == 0:
        li1 = []
    else:
        li1 = li1.split(", ")
    for i in range(len(li1)):
        li1[i] = int(li1[i])
    li2 = input("B:")[1:-1]
    if len(li2) == 0:
        li2 = []
    else:
        li2 = li2.split(", ")
    for i in range(len(li2)):
        li2[i] = int(li2[i])
wholeSet = {1: union(li1, li2), 2: intersection(li1, li2), 3: complement(li2, li1), 4: difference(li1, li2), 5: crossProduct(tuple(li1), tuple(li2))}
print(wholeSet[setOperation])
```
