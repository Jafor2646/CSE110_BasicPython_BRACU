```python
#Task23
def splitting_money(amount):
  store = ""
  note500 = amount // 500
  if note500 != 0:
    store += f"500 Taka: {note500} note(s)\n" 
  amount %= 500
  note100 = amount // 100
  if note100 != 0:
    store += f"100 Taka: {note100} note(s)\n" 
  amount %= 100
  note50 = amount // 50
  if note50 != 0:
    store += f"50 Taka: {note50} note(s)\n" 
  amount %= 50
  note20 = amount // 20
  if note20 != 0:
    store += f"20 Taka: {note20} note(s)\n" 
  amount %= 20
  note10 = amount // 10
  if note10 != 0:
    store += f"10 Taka: {note10} note(s)\n" 
  amount %= 10
  note5 = amount // 5
  if note5 != 0:
    store += f"5 Taka: {note5} note(s)\n" 
  amount %= 5
  note2 = amount // 2
  if note2 != 0:
    store += f"2 Taka: {note2} note(s)\n" 
  amount %= 2
  if amount != 0:
    store += f"1 Taka: {amount} note(s)"
  return store
print(splitting_money(1234)) 
```

```python
#Task25
def function_name(sv,ev,fd,ed):
  sum = 0
  for i in range(sv,ev) :
    if i % fd == 0 and i % ed != 0:
      sum += i
    elif i % ed == 0 and i % fd != 0:
      sum += i
  return sum
startingValue = int(input("Enter starting value:"))
endingValue = int(input("Enter ending value:"))
firstDivisor = int(input("Enter first divisor:"))
secondDivisor = int(input("Enter second divisor:"))
print(function_name(startingValue,endingValue,firstDivisor,secondDivisor))  
```
