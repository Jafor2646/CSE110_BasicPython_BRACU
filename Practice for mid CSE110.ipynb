```python
var = input("Enter something:")
end = 'stop'
product = int(var)
if var in end:
  print(0)
while var not in end:
  var = input("Enter something:")
  if var == end:
    print(product)
  else:
    var1 = int(var)
    product *= var1
```

```python
var=input("Enter sentence:")
count1 = 0
count2 = 0
count3 = 0
for i in var:
  if ord(i) >= 97 and ord(i) <= 122:
    if ord(i) == 97 or ord(i) == 101 or ord(i) == 105 or ord(i) == 111 or ord(i) == 117:
      count1 += 1
    else:
      count2 += 1
  if i=='0' or i=='1' or i=='2' or i=='3' or i=='4' or i=='5' or i=='6' or i=='7' or i=='8' or i=='9':
    count3 += 1
print("Vowels:", count1)
print("Consonants:", count2)
print("Digits:", count3)
```
