```python
#Task 25
var1 = input("Enter sentence:").split(".")[:-1]
var2 = input("Enter symbols:")[1:-1].split(", ")
outputDict = {}
symbolList = []
wordList = []
for i in var2:
  symbolList.append(i[1])
for i in range(len(var1)):
  temp = var1[i].strip()
  temp = temp.split(" ")
  for j in temp:
    wordList.append(j)
for i in range(len(wordList)):
  string = wordList[i]
  for j in range(len(string)):
    temp = 0
    temp += ord(string[j])
  calculatedIndex = temp % len(symbolList)
  print(temp)
  print(calculatedIndex)
  if symbolList[calculatedIndex] not in outputDict:  
    outputDict[symbolList[calculatedIndex]] = [wordList[i]]
  else:
    if wordList[i] not in outputDict.values():
      outputDict[symbolList[calculatedIndex]].append(wordList[i])
print(outputDict)

```

```python
def productOfList(myDict):
  myList = []
  mul = 1
  for k,v in myDict.items():
    x,y = tuple(v)
    mul = ord(x)*ord(y)
    myList.append(mul)
  return tuple(myList)
print(productOfList({"A":['a','b'],"B":['g','z'],"C":['d','h'],"D":['q','r']}))
```
