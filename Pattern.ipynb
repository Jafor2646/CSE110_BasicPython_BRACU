```python
#right triagle right justified(first increasing then decreasing)
var = int(input("Enter a number:"))
increase = var

for row in range(1, var+1):
  for s in range(row+1):
    print(" ",end="")
  for col in range(increase,0,-1):
    print(col,end="")
  print()
  increase -= 1
```
